### Visualising Education Level On Matplotlib Plot

```python
data['EdLevel'].value_counts(normalize=True) * 100
```

```python
from matplotlib import pyplot as plt
percent_spread = data['EdLevel'].value_counts(normalize=True) * 100
percent_spread.sort_values()
percent_spread.plot(kind='barh')
plt.title('Distribution of education levels')
plt.xlabel('Percentage')
plt.ylabel("Education Level")

plt.show()
```

### Calculating offline learning percentage share

```python
learn_methods = data['LearnCode'].str.split(";").explode()
learn_methods.value_counts(normalize=True) * 100
```

```python
method_count = learn_methods.value_counts()
method_percentage = (method_count / 89184) * 100
method_percentage
```

```python
method_percentage.plot(kind='barh')
plt.title('Learning methods')
plt.xlabel('Percentage')
plt.ylabel('Learn Code')
for index,value in enumerate(method_percentage):
    plt.text(value,index,f'{round(value,2)}')
plt.show()
```
